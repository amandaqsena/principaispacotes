```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("dados/dadosPerformance.txt")
df.head()
df["Iteracoes"] = df["Iteracoes"].astype(str)
```

```python
df['ratioAdicionarll/al'] = df["adicionarAL"]/df["adicionarLL"]
df['ratioremoverll/al'] = df["removerAL"]/df["removerLL"]
```

```python
plt.plot(df["Iteracoes"],df["ratioAdicionarll/al"])
plt.plot(df["Iteracoes"],df["ratioremoverll/al"])
```

```python
labels = df["Iteracoes"]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(2,1,figsize=(6,8))

rects1 = ax[0].bar(x - width/2, df["adicionarAL"], width, label='Adicionar à Array List')
rects2 = ax[0].bar(x + width/2, df["adicionarLL"], width, label='Adicionar à Linked List')


# Add some text for labels, title and custom x-axis tick labels, etc.
ax[0].set_ylabel('Tempo')
ax[0].set_xticks(x, labels)
ax[0].legend()


rects3 = ax[1].bar(x - width/2, df["removerAL"], width, label='Remover da Array List')
rects4 = ax[1].bar(x + width/2, df["removerLL"], width, label='Remover da Linked List')


# Add some text for labels, title and custom x-axis tick labels, etc.
ax[1].set_ylabel('Tempo')
ax[1].set_xticks(x, labels)
ax[1].legend()


fig.tight_layout()

plt.show()
```
